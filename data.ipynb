# Inflationrate and Unemploymentrate trend in united states from 1980 - 2020


```python
co=(df["country"]=="United States of America")
```

```python
df1=df.loc[co,["inflationrate","unemploymentrate","year"]].set_index("year")
```

```python
df1
```

```python



import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df1.plot(kind='line')

plt.title('inflationrate and unemploymentrate trend in united states from 1980 - 2020')

plt.xlabel('Years')

plt.show()
```

## Conclusion :
### Inflation rate is highest in 1980 for us and unemploymenrate is highest in 2010

# analayising percapitaincome of countries over the years

```python
gb=df.groupby(["country"])
```

```python
gb.get_group("India")
```

```python
pe=gb["percapitaincome"].agg(['mean','median'])
```

```python
pe
```

```python
pe.plot(kind='line',linewidth=3,linestyle='solid',marker='o',markerfacecolor='red',color=['y',"g"],figsize=(14,12))
plt.title('mean and median of percapitaincome of countries over 1980-2020')
plt.xlabel("countries")
plt.ylabel("percapitaincome")
plt.xticks(rotation=90)
plt.show()
```
